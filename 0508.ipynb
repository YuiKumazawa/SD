```python
from matplotlib import pyplot as plt

x = [1,2,4,8,16,32,64,128,256]
#xs = [i for i,_ in enumerate(x)]
#xs=[]for i in range(0,len(x)): xs.append(i)

xs=[]
for i in range(0,len(x)):
    xs.append((i+1)*100)

plt.plot(xs,x,'-g',label='variance')

plt.legend(loc=9)

plt.show()
```

```python
from matplotlib import pyplot as plt

x = [1,2,4,8,16,32,64,128,256]
y = []
for i in range(len(x)-1,-1,-1):
    y.append(x[i])
z = []
for i in range(0,len(x)):
    z.append(x[i]+y[i])
xs=[]
for i in range(0,len(x)):
    xs.append((i+1)*100)

plt.plot(xs,x,'-g',label='variance')
plt.plot(xs,y,'-.r',label='bias^2')
plt.plot(xs,z,':b',label='total error')

plt.legend(loc=9)

plt.show()
```

```python
from matplotlib import pyplot as plt

x = [70,65,72,63,71,64]
y = [175,170,205,120,220,130]

plt.scatter(x,y)

for i in range(0,len(x)):
    label = "[" + str(x[i]) + "," + str(y[i]) + "]"
    plt.annotate(label, xy=(x[i],y[i]), xytext=(5,-5), textcoords='offset points')

plt.show()
```
